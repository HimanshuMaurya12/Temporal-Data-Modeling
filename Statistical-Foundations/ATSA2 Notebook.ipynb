```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller

# --- Prerequisite: Generate Stationary Series C from Lab 1 ---
np.random.seed(42)
series_C = np.random.normal(0, 1, 100)

# --- 1. Plot ACF of Stationary Data ---
# This shows how a value correlates with its past values (lags)
plt.figure(figsize=(10, 5))
plot_acf(series_C, lags=20)
plt.title("ACF of Stationary Series C")
plt.xlabel("Lags")
plt.ylabel("Autocorrelation")
plt.grid(True, alpha=0.3)
plt.show()

# --- 2. Perform Augmented Dickey-Fuller (ADF) Test ---
# This is a formal statistical test for stationarity
print("--- Augmented Dickey-Fuller Test Results ---")
adf_result = adfuller(series_C)

print(f"ADF Statistic: {adf_result[0]:.4f}")
print(f"p-value: {adf_result[1]:.4f}")

# Interpretation logic
if adf_result[1] < 0.05:
    print("Result: Series is Stationary (Reject Null Hypothesis)")
else:
    print("Result: Series is NOT Stationary (Fail to Reject Null Hypothesis)")

# --- 3. Comparison (Optional but recommended for Lab 2) ---
# Generating a trending series to see the difference in ACF
time = np.arange(100)
series_A_trend = 0.5 * time + np.random.normal(0, 2, 100)

plt.figure(figsize=(10, 5))
plot_acf(series_A_trend, lags=20)
plt.title("ACF of Trending Series (For Comparison)")
plt.show()
```
