```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# --- 1. Create Time Series ---
time = np.arange(100)

# Series A: Trend (Upward trend)
series_A = 0.5 * time + np.random.normal(0, 2, 100)

# Series B: Seasonal (Clear repeating pattern)
series_B = 10 + 5 * np.sin(2 * np.pi * time / 12) + np.random.normal(0, 1, 100)

# Series C: Stationary (No trend or seasonality)
series_C = np.random.normal(0, 1, 100)
```

```python
# --- 2. Plot all 3 series ---
plt.figure(figsize=(12, 8))

# Plot Trend
plt.subplot(3, 1, 1)
plt.plot(series_A, color='blue')
plt.title("Series A - Trend")

# Plot Seasonality
plt.subplot(3, 1, 2)
plt.plot(series_B, color='green')
plt.title("Series B - Seasonality")

# Plot Stationary
plt.subplot(3, 1, 3)
plt.plot(series_C, color='red')
plt.title("Series C - Stationary")

plt.tight_layout()
plt.show()
```

```python
# --- 3. Compute Mean & Variance ---
data = {'Series': ['Trend (A)', 'Seasonal (B)', 'Stationary (C)'],
        'Mean': [np.mean(series_A), np.mean(series_B), np.mean(series_C)],
        'Variance': [np.var(series_A), np.var(series_B), np.var(series_C)]}

df_stats = pd.DataFrame(data)
print(df_stats)
```
